## Protein in ionic liquids

```python
#Load Stuff
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
from pylab import rcParams

# import seaborn to make nice figures.
import seaborn as sns

rcParams['figure.figsize'] = 10, 5
rcParams['lines.linewidth'] = 2
rcParams['font.size'] = 10
```

```python
#%matplotlib notebook

plt.close()
data=pd.read_csv('3AMD_310_nc.all.res.dat', delim_whitespace=True)
result = pd.DataFrame(data, columns=['#Frame'])

for system in ['3AMD','3AXX'] :
    l=[]
    for temp in ['310','353'] :
        data=pd.read_csv('{}_{}_nc.all.res.dat'.format(system,temp), delim_whitespace=True)
        df = pd.DataFrame(data, columns=['NC1[native]'])
        df=df.rename(columns={"NC1[native]": system+"_"+temp})
        result = pd.concat([result, df], axis=1, join='inner')
        l.append(system+"_"+temp)

        plt.plot(df,alpha=0.2)
        plt.plot(pd.rolling_mean(df,50),label=system+"_"+temp)
        
    plt.xlabel("Frame")
    plt.ylabel(system+" Native contacts")
    plt.legend()
    plt.show()

    result.boxplot(l)
    plt.xlabel("System")
    plt.ylabel(system+" Native contacts")
    plt.legend()
    plt.show()

plt.title("Diferença nos contatos nativos (310 - 353)")
d3amd=pd.rolling_mean(result['3AMD_310']-result['3AMD_353'], 50)
d3axx=pd.rolling_mean(result['3AXX_310']-result['3AXX_353'], 50)
plt.plot(d3amd, label='3AMD')
plt.plot(d3axx, label='3AXX')
plt.legend()
plt.show()

plt.bar(0,np.mean(d3amd),yerr=np.std(d3amd))
plt.bar(1,np.mean(d3axx),yerr=np.std(d3axx))
plt.legend()
plt.show()

```

```python
# Plot Histogram of PC1 and PC2 projections
for system in ['3AMD','3AXX'] :
    for temp in ['310','353'] :
        data = np.genfromtxt('{}_{}_project.dat'.format(system,temp))
#        plt.hist2d(data[:,1],data[:,2],bins=100)
        plt.scatter(data[:,1],data[:,2], c=data[:,0]*0.005,cmap='jet', lw = 0)

        # Set title and axis names
        plt.title("Projection along Principal Components")
        plt.xlabel("PC1")
        plt.ylabel(system+"_"+temp+"\n\nPC2")

        cbar = plt.colorbar()
        cbar.ax.set_ylabel('Time (ns)')

        window=2000
        maxwindow=100000
        x=[]
        y=[]
        for i in range(0, maxwindow,window) :
            x.append(np.mean(data[i:window+i,1]))
            y.append(np.mean(data[i:window+i,2]))

        x=np.asarray(x)
        y=np.asarray(y)

        #plt.plot(x,y, color='white')

        plt.quiver(x[:-1], y[:-1], x[1:]-x[:-1], y[1:]-y[:-1], scale_units='xy', angles='xy', scale=1, color='white', linewidths=(0.5))

        plt.xlim(-35,35)
        plt.ylim(-35,35)
        plt.show()
    
```
